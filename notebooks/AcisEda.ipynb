```python
# Checking the number of rows and columsn
rows,cols = acis_data.shape
print(f"Number of rows:{rows}")
print(f"Number of columns:{cols}")
```

```python
#Handling Missing Values
acis_data['TotalClaims'].fillna(0)
# Binary Flags
binary_cols = [
    'IsVATRegistered', 'AlarmImmobiliser', 'TrackingDevice',
    'NewVehicle'
]
acis_data[binary_cols] = acis_data[binary_cols].fillna(False)
# Categorical columns
categorical_cols = [
    'Citizenship', 'LegalType', 'Title', 'Language', 'Bank',
    'AccountType', 'MaritalStatus', 'Gender', 'Country',
    'Province', 'MainCrestaZone', 'SubCrestaZone',
    'ItemType', 'VehicleType',  'Model', 
    'CoverCategory', 'CoverType', 'CoverGroup', 'Section',
    'Product', 'StatutoryClass', 'StatutoryRiskType',
    'TermFrequency', 'ExcessSelected'
]
acis_data[categorical_cols] = acis_data[categorical_cols].fillna('unknown')

# drop rows with missing premium
acis_data = acis_data.dropna(subset=['TotalPremium'])
```

```python
# Additional Missing Values Cleaning
acis_data['mmcode']=acis_data['mmcode'].fillna('unknown')
acis_data['make']=acis_data['make'].fillna('unknown')
acis_data['bodytype']=acis_data['bodytype'].fillna('unknown')

for col in ['cubiccapacity', 'kilowatts']:
    acis_data[col] = acis_data.groupby('VehicleType')[col].transform(
        lambda x: x.fillna(x.median())
    )
    acis_data[col] = acis_data[col].fillna(acis_data[col].median())
```

```python
# Missing value cleaning continued
acis_data['VehicleIntroDate']=acis_data['VehicleIntroDate'].fillna(1)
acis_data['CapitalOutstanding']=acis_data['CapitalOutstanding'].fillna('Unknown')
```

```python
# Missing value cleaning continued
for col in ['Cylinders', 'NumberOfDoors']:
    acis_data[col] = acis_data.groupby('VehicleType')[col].transform(
        lambda x: x.fillna(x.median())
    )
    acis_data[col] = acis_data[col].fillna(acis_data[col].median()) 
```

```python
# Missing value cleaning continued
acis_data['CustomValueEstimate']=acis_data['CustomValueEstimate'].fillna(acis_data['SumInsured'])
acis_data['CustomValueMissing']=(acis_data['CustomValueEstimate'].isna()).astype(int)
status_cols = ['WrittenOff', 'Rebuilt', 'Converted']
acis_data[status_cols]=acis_data[status_cols].fillna(False)
acis_data['CrossBorder'] = acis_data['CrossBorder'].fillna(False)
acis_data['NumberOfVehiclesInFleet'] = acis_data['NumberOfVehiclesInFleet'].fillna(1)


```

```python
#Counting Missing Values
missing_values = acis_data.isnull().sum()
print(missing_values) 
```

```python
# shape count after Missing Value Cleaning
rows,cols = acis_data.shape
print(f"Number of Rows:{rows}")
print(f"Number of columns:{cols}")
```

```python
#Feature Enginnering
#Time based features
acis_data['TransactionMonth']=pd.to_datetime(acis_data['TransactionMonth'])
acis_data['TransactionYear']=acis_data['TransactionMonth'].dt.year
acis_data['TransactionMonthNum']=acis_data['TransactionMonth'].dt.month
# Vechile Age
acis_data['VehicleAge'] = acis_data['TransactionYear']-acis_data['TransactionYear']
acis_data['VehicleAge'] = acis_data['VehicleAge'].clip(lower=0)

# Engine Power Ratios
acis_data['PowerToCapacity'] = acis_data['kilowatts'] / acis_data['cubiccapacity']

# Financial Risk & Exposure Features
acis_data['SumInsuredToValue'] = acis_data['SumInsured'] / (acis_data['CustomValueEstimate'] + 1)

```

```python
# Claim History Proxies
acis_data['HasClaim'] = (acis_data['TotalClaims'] > 0).astype(int)
acis_data['LossRatio'] = acis_data['TotalClaims'] / (acis_data['TotalPremium'] + 1)
```

```python
# Columns to be excluded from Encoding
exclude_cols = [
    'UnderwrittenCoverID',
    'PolicyID',
    'TransactionMonth',
    'VehicleIntroDate'
]
```

```python
# Binary/boolean columns:-Integer encoding
binary_cols = [
    'IsVATRegistered', 'AlarmImmobiliser', 'TrackingDevice',
    'NewVehicle', 'WrittenOff', 'Rebuilt', 'Converted', 'CrossBorder'
]

acis_data[binary_cols] = acis_data[binary_cols].apply(
    lambda x: x.astype(str).str.lower().map({'true': 1, 'false': 0})
)
```

```python
low_cardinality = [
    'LegalType', 'Title', 'Language', 'AccountType',
    'MaritalStatus', 'Gender', 'Country', 'Province',
    'VehicleType', 'bodytype', 'TermFrequency',
    'CoverCategory', 'CoverType', 'CoverGroup',
    'Section', 'Product', 'StatutoryClass', 'StatutoryRiskType'
]
# Encoding
acis_data = pd.get_dummies(
    acis_data,
    columns=low_cardinality,
    drop_first=True
)
```

```python
# Frequency Encoing-Leakage-Safe
high_card_cols = [
    'make', 'Model', 'mmcode', 'Bank',
    'PostalCode', 'MainCrestaZone', 'SubCrestaZone'
]

for col in high_card_cols:
    freq = acis_data[col].value_counts() / len(acis_data)
    acis_data[col + '_freq'] = acis_data[col].map(freq)
```

```python
# Columns to keep as numeric
numeric_cols = [
    'RegistrationYear', 'Cylinders', 'cubiccapacity', 'kilowatts',
    'NumberOfDoors', 'CustomValueEstimate', 'CapitalOutstanding',
    'NumberOfVehiclesInFleet', 'SumInsured', 'CalculatedPremiumPerTerm',
    'TransactionYear', 'TransactionMonthNum', 'VehicleAge',
    'PowerToCapacity', 'SumInsuredToValue', 'CustomValueMissing'
]
```

```python
# Drop Original High-Cardinality Columns (After Encoding)
acis_data.drop(columns=high_card_cols, inplace=True)
```

```python
acis_data.drop(columns=['VehicleIntroDate'], inplace=True)
```

```python
acis_data['CapitalOutstanding'] = pd.to_numeric(
    acis_data['CapitalOutstanding'], errors='coerce'
)
```

```python
acis_data['CapitalOutstanding'] = acis_data['CapitalOutstanding'].fillna(
acis_data['CapitalOutstanding'].median()
)
```

```python
#citizenship - Low Cardinality - onehot encoding
acis_data = pd.get_dummies(
    acis_data,
    columns=['Citizenship'],
    drop_first=True
)
```

```python
#ItemType - Low Cardinality - onehot encoding
acis_data = pd.get_dummies(
    acis_data,
    columns=['ItemType'],
    drop_first=True
)
```

```python
#ExcessSelected-Ordinal Numeric Like
acis_data['ExcessSelected'] = (
    acis_data['ExcessSelected']
    .astype(str)
    .str.replace('[^0-9.]', '', regex=True)
)

acis_data['ExcessSelected'] = pd.to_numeric(
    acis_data['ExcessSelected'], errors='coerce'
)

acis_data['ExcessSelected'] = acis_data['ExcessSelected'].fillna(
    acis_data['ExcessSelected'].median()
)
```

```python
# Columns to exclude from x
exclude_from_X = [
    'UnderwrittenCoverID',
    'PolicyID',
    'TotalClaims',
    'TotalPremium',
    'LossRatio',
    'MarginRatio'
]
```

```python
# claims model dataset
x_claims= acis_data.drop(columns=exclude_from_X)
y_claims=acis_data['TotalClaims']
```

```python
# premium model dataset
x_premium=acis_data.drop(columns=exclude_from_X)
y_premium=acis_data['TotalPremium']
```

```python
from sklearn.model_selection import train_test_split
X_claims_train, X_claims_test, y_claims_train, y_claims_test = train_test_split(
    x_claims, y_claims, test_size=0.2, random_state=42
)

X_premium_train, X_premium_test, y_premium_train, y_premium_test = train_test_split(
    x_premium, y_premium, test_size=0.2, random_state=42
)
```

```python
# Droping TransactionMonth Column
X_claims_train = X_claims_train.drop(columns=['TransactionMonth'])
X_claims_test  = X_claims_test.drop(columns=['TransactionMonth'])

X_premium_train = X_premium_train.drop(columns=['TransactionMonth'])
X_premium_test  = X_premium_test.drop(columns=['TransactionMonth'])
```

```python
# Droping columns with two many Missing Values
cols_to_drop = ['TrackingDevice', 'AlarmImmobiliser']
X_claims_train = X_claims_train.drop(columns=cols_to_drop)
X_claims_test  = X_claims_test.drop(columns=cols_to_drop)

X_premium_train = X_premium_train.drop(columns=cols_to_drop)
X_premium_test  = X_premium_test.drop(columns=cols_to_drop)
```

```python
from sklearn.impute import SimpleImputer

# Boolean / categorical numeric â†’ fill with 0
bool_like_cols = ['NewVehicle', 'Rebuilt', 'Converted', 'WrittenOff', 'CrossBorder']
for col in bool_like_cols:
    X_claims_train[col].fillna(0, inplace=True)
    X_claims_test[col].fillna(0, inplace=True)
    X_premium_train[col].fillna(0, inplace=True)
    X_premium_test[col].fillna(0, inplace=True)

# Numeric features â†’ median
num_cols = ['PowerToCapacity']
X_claims_train = X_claims_train.drop(columns=num_cols)
X_claims_test  = X_claims_test.drop(columns=num_cols)

X_premium_train = X_premium_train.drop(columns=num_cols)
X_premium_test  = X_premium_test.drop(columns=num_cols)
```

```python
# Modeling: Linear Regression claims Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
#Train the Model
lr_claims=LinearRegression(n_jobs=-1)
lr_claims.fit(X_claims_train,y_claims_train)
```

```python
# Predict on the test set
y_claims_pred = lr_claims.predict(X_claims_test)
# Evaluate Performance
rmse_claims = np.sqrt(mean_squared_error(y_claims_test, y_claims_pred))
mae_claims  = mean_absolute_error(y_claims_test, y_claims_pred)
r2_claims   = r2_score(y_claims_test, y_claims_pred)
print("Linear Regression â€“ Claims Model")
print(f"RMSE: {rmse_claims:.4f}")
print(f"MAE : {mae_claims:.4f}")
print(f"RÂ²  : {r2_claims:.4f}")
```

```python
#Linear Regression: Premium Model Prediction
lr_premium = LinearRegression(n_jobs=-1)
lr_premium.fit(X_premium_train, y_premium_train)
# Predict
y_premium_pred = lr_premium.predict(X_premium_test)
# Evaluation
rmse_premium = np.sqrt(mean_squared_error(y_premium_test, y_premium_pred))
mae_premium  = mean_absolute_error(y_premium_test, y_premium_pred)
r2_premium   = r2_score(y_premium_test, y_premium_pred)

print("Linear Regression â€“ Premium Model")
print(f"RMSE: {rmse_premium:.4f}")
print(f"MAE : {mae_premium:.4f}")
print(f"RÂ²  : {r2_premium:.4f}")
```
