```python
# extracting 2 Numeric and 2 categorical Columns for the data visualiztion
acis_numeric = ['PolicyID','PostalCode']
acis_category = ['MaritalStatus','Province']
```

```python
for col in acis_numeric:
    plt.figure(figsize=(4,2))
    sns.histplot(acis_data[col],kde=True, bins=5)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Bar charts for categorical columns
for col in acis_category:
    plt.Figure(figsize=(4,2))
    sns.countplot(data=acis_data,x=col,order=acis_data[col].value_counts().index)
    plt.title(f"frequency of {col}")
    plt.xticks(rotation=90) #rotate label if too many categories
    plt.show()
```

```python
# Bivariate Analysis
plt.figure(figsize=(8,6))
sns.scatterplot(data=acis_data,x='TotalPremium',y='TotalClaims',hue='PostalCode')
plt.title("Relationship between Total Premium and Total Claim by ZipCode")
plt.show()
```

```python
# Correlation Matrix 
corr = acis_data[['TotalPremium','TotalClaims']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation between TotalPremium and TotalClaims")
plt.show()
```
